```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
coffeeProd = pd.read_csv('coffee-bean-production.csv')

#removing stange labeling of 'Entities' and the long label for Produce
coffeeProd.columns = ['Country', 'Code', 'Year', 'Produce']

brazil = coffeeProd[coffeeProd['Country'] == 'Brazil']
china = coffeeProd[coffeeProd['Country'] == 'China']
india = coffeeProd[coffeeProd['Country'] == 'India']
usa = coffeeProd[coffeeProd['Country'] == 'United States']

print(coffeeProd)
```

```python
fig, ax = plt.subplots(figsize=(17,12))

#plotting lines
ax.plot(brazil['Year'], brazil['Produce'], color = 'red', marker = 'o')
ax.plot(china['Year'], china['Produce'], label = 'China', color = 'green', marker = 'o')
ax.plot(india['Year'], india['Produce'], label = 'India', color = 'orange', marker = 'o')
ax.plot(usa['Year'], usa['Produce'], label = 'United States', color = 'purple', marker = 'o')

#setting axis grid, limits, and tick labels, labeling each line
ax.axis([1961, 2021, 0, 3500000])
ax.grid(axis = 'y', linestyle = '--',)
ax.annotate('Brazil', xy=(2021,3000000), xytext=(2022, 3050000), color='red', fontsize=15, \
            arrowprops=dict(arrowstyle='-'))
ax.annotate('India', xy=(2021,300000), xytext=(2022, 350000), color='orange', fontsize=15, \
            arrowprops=dict(arrowstyle='-'))
ax.annotate('China', xy=(2021,115000), xytext=(2022, 119000), color='green', fontsize=15, \
            arrowprops=dict(arrowstyle='-'))
ax.annotate('United States', xy=(2021,5000), xytext=(2022, 7000), color='purple', fontsize=15, \
            arrowprops=dict(arrowstyle='-'))
ax.set_xticks([1961, 1970, 1980, 1990, 2000, 2010, 2021])
ax.set_yticks([0, 500000, 1000000, 1500000, 2000000, 2500000, 3000000, 3500000], ['0 t', \
              '500,000 t', '1 million t', '1.5 million t', '2 million t', '2.5 million t', \
              '3 million t', '3.5 million t'])

plt.suptitle('Coffee bean production, 1961 to 2021, (Recreation in Matplotlib/pyplot w/ Pandas)',\
            fontsize=25)
plt.title('Coffee bean production is measured in tonnes.', fontsize=20, loc='left')
plt.xlabel('Created by Trent Hogarth, original by Our World In Data', loc='left')

plt.show()
```
